```python
import json
import pandas as pd
data = open('hyperparams_data')
```

```python
dict_1={}
i=0
for line in data:
    #print(line)
    res = json.loads(line)
    dict_1[i]=res
    i=i+1
    
df = pd.DataFrame(dict_1).transpose()
```

```python
df
```

```python
def lst_hp(dict_hp):
    lst_hp=[]
    for key,values in dict_hp.items():
        lst_hp.append(key)
    return lst_hp
```

```python
hp_dict = {}
for i in df["hyperparameters"]:
    for j in i:
        if j not in hp_dict:
            hp_dict[j] = 1
        else:
            hp_dict[j] += 1
            
hp_dict
```

```python

for i in lst_hp(hp_dict):
    df[i] = df['hyperparameters'].apply(lambda x: x.get(i))
   
df
```

```python
df['Lst_HP']=df['hyperparameters'].apply(lst_hp)
```

```python
df.drop(['hyperparameters'],axis=1,inplace=True)
```

```python
df
```

```python
data = open('features_data')
```

```python
dict_2={}
i=0
for line in data:
    #print(line)
    res = json.loads(line)
    dict_2[i]=res
    i=i+1
    
df_1= pd.DataFrame(dict_2).transpose()
```

```python
df_1
```

```python
df_1['features']=df_1["feature_names"].apply(lambda x: len(x))
```

```python
df_1['features'].max()
```

```python
def to_1D(series):
 return pd.Series([x for _list in series for x in _list])
```

```python
feature_lst = to_1D(df_1["feature_names"]).tolist()
```

```python
feature_lst
```

```python
def boolean_df(item_lists, unique_items):
# Create empty dict
    bool_dict = {}
    
    # Loop through all the tags
    for item in unique_items:
        
        # Apply boolean mask
        bool_dict[item] = item_lists.apply(lambda x: item in x)
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_dict)
```

```python
features=boolean_df(df_1["feature_names"],feature_lst)
```

```python
Features=pd.concat([df_1,features],axis=1)
```

```python
Features.replace(to_replace=True,
    value=1,
    inplace=True)
Features.replace(to_replace=False,
    value=0,
    inplace=True)
```

```python
Features.head()
```

```python
data = open('study_level_data')
dict_1={}
i=0
for line in data:
    #print(line)
    res = json.loads(line)
    dict_1[i]=res
    i=i+1
    
Study = pd.DataFrame(dict_1).transpose()
Study
```

```python
Cross_study_1=pd.merge(df,Features,how='left',left_on='parent_study_id',
    right_on='study_id')
Cross_study_1
```

```python
Cross_study_Final=pd.merge(Cross_study_1,Study,how='left',left_on='parent_study_id',
    right_on='study_id')
Cross_study_Final
```
